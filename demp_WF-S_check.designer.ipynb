```python
"""
id: combine_41
template: combine
templateVersion: 2.0.0
name: combine_41
position:
  x: 2700
  y: 586.25
description:
  text: Combine two datasets into one by matching columns by name and removing duplicates.
  hash: a78afeb1
previewMode: "1000"
config:
  operator: UNION
  quantifier: DISTINCT
  union_by_name: true
  allow_missing_columns: false
input:
  - node: join_20
    input_port: data
    output_port: joined_data
  - node: join_38_1_1
    input_port: data
    output_port: joined_data
"""

# generated from the system
from functools import reduce
from typing import Any, Callable, Dict, List

def run(
    config: Dict[str, Any], inputs: Dict[str, Any], spark
) -> Dict[str, Any]:
    dfs: List[Any] = list(inputs.get("data") or [])
    if len(dfs) < 2:
        raise ValueError(
            f"Combine requires at least 2 inputs, got {len(dfs)}. "
            "Wire 2 or more upstream operators into the 'data' port."
        )

    operator = config.get("operator", "UNION")
    quantifier = config.get("quantifier", "DISTINCT")
    union_by_name = bool(config.get("union_by_name", False))
    allow_missing_columns = bool(config.get("allow_missing_columns", False))

    pair: Callable[[Any, Any], Any]
    if operator == "UNION":
        if union_by_name:
            def pair(a: Any, b: Any) -> Any:
                return a.unionByName(b, allowMissingColumns=allow_missing_columns)
        else:
            def pair(a: Any, b: Any) -> Any:
                return a.union(b)
    elif operator == "INTERSECT":
        if quantifier == "ALL":
            def pair(a: Any, b: Any) -> Any:
                return a.intersectAll(b)
        else:
            def pair(a: Any, b: Any) -> Any:
                return a.intersect(b)
    elif operator in ("EXCEPT", "MINUS"):
        if quantifier == "ALL":
            def pair(a: Any, b: Any) -> Any:
                return a.exceptAll(b)
        else:
            def pair(a: Any, b: Any) -> Any:
                return a.subtract(b)
    else:
        raise ValueError(f"Unsupported combine operator: {operator}")

    result = reduce(pair, dfs)
    if operator == "UNION" and quantifier == "DISTINCT":
        result = result.distinct()
    return {"combined_data": result}

# generated from the system
ctx = globals().setdefault("ctx", {})
config = {
    "operator": "UNION",
    "quantifier": "DISTINCT",
    "union_by_name": True,
    "allow_missing_columns": False
}
inputs = {
    "data": [
        ctx["join_20.joined_data"],
        ctx["join_38_1_1.joined_data"]
    ],
    "data__sources": [
        {
            "node": "join_20",
            "output_port": "joined_data",
            "name": "join_20",
            "df_name": "join_20"
        },
        {
            "node": "join_38_1_1",
            "output_port": "joined_data",
            "name": "join_38_copy_copy",
            "df_name": "join_38_copy_copy"
        }
    ]
}
out = run(config, inputs, spark)
ctx["combine_41.combined_data"] = out["combined_data"]
```
